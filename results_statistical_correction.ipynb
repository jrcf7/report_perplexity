```python
import pandas as pd
import pingouin as pg
```

```python
df_names = [
    'DB_randperm_peacock_OLMo-7B.csv',
    'DB_randperm_peacock_GPT2.csv',
    'WPvsDB_peacock_test_results.csv',
    'WPvsDB_rand_perm_results.csv',
    'WPvsDB_pplnwd_corr_results.csv'
]
```

```python
dfs = []
for d_n in df_names:
    dfs.append(pd.read_csv(d_n))

p_val_df = pd.concat(dfs).reset_index(drop=True)
```

```python
reject, pvals_corr = pg.multicomp(list(p_val_df.p), method='holm')
```

```python
p_val_df["Corrected p"] = pvals_corr
p_val_df["Still significant"] = reject
```

```python
p_val_df.to_csv('stats_corrected_all.csv', index=False)
```
